```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score

def plot_confusion_matrix(test_y, predict_y):
    C = confusion_matrix(test_y, predict_y)    
    A =(((C.T)/(C.sum(axis=1))).T)    
    B =(C/C.sum(axis=0))
    plt.figure(figsize=(20,4))
    
    labels = [0,1]
    # representing A in heatmap format
    cmap=sns.light_palette("green")
    plt.subplot(1, 3, 1)
    sns.heatmap(C, annot=True, cmap=cmap, fmt=".3f", xticklabels=labels, yticklabels=labels)
    plt.xlabel('Predicted Class')
    plt.ylabel('Original Class')
    plt.title("Confusion matrix")
    
    plt.subplot(1, 3, 2)
    sns.heatmap(B, annot=True, cmap=cmap, fmt=".3f", xticklabels=labels, yticklabels=labels)
    plt.xlabel('Predicted Class')
    plt.ylabel('Original Class')
    plt.title("Precision matrix")
    
    plt.subplot(1, 3, 3)
    # representing B in heatmap format
    sns.heatmap(A, annot=True, cmap=cmap, fmt=".3f", xticklabels=labels, yticklabels=labels)
    plt.xlabel('Predicted Class')
    plt.ylabel('Original Class')
    plt.title("Recall matrix")
    
    plt.show()
```

```python
import pandas as pd
from patsy import dmatrices
from sklearn.model_selection import train_test_split

df = pd.read_csv('crowd.csv')
col = df.columns.values.tolist()
col.remove('y')
col.remove('votes')
col.remove('comments')

x = "+".join(col)
y, X = dmatrices('y ~' + x, df, return_type='dataframe')

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
y_test.y.value_counts()
```

```python
from sklearn.linear_model import RidgeClassifierCV

ridge = RidgeClassifierCV(cv=10)
ridge.fit(x_train, y_train.values.ravel())
ridge_predict = ridge.predict(x_test) 
print(accuracy_score(y_test, ridge_predict))  
print("Ridge AUC : ", roc_auc_score(y_test, ridge_predict))
plot_confusion_matrix(y_test, ridge_predict) 
```

```python
from sklearn.linear_model import LogisticRegression

# penalty='l1' indicates it is lasso
logistic_regression = LogisticRegression(penalty='l1')
logistic_regression.fit(x_train, y_train.values.ravel())
predictions = logistic_regression.predict(x_test)
print(accuracy_score(y_test, predictions)) 
print("Lasso AUC : ", roc_auc_score(y_test, predictions))
plot_confusion_matrix(y_test, predictions)
```

```python
from pygam import LogisticGAM

gam = LogisticGAM().fit(x_train, y_train.values.ravel())
predict_gam = gam.predict(x_test)
print(accuracy_score(y_test, predict_gam))
print("GAM AUC : ", roc_auc_score(y_test, predict_gam))
plot_confusion_matrix(y_test, predict_gam)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(random_state=0)
dtc.fit(x_train, y_train.values.ravel())
predict_dtc = dtc.predict(x_test)
print(accuracy_score(y_test, predict_dtc))
print("Trees AUC : ", roc_auc_score(y_test, predict_dtc))
plot_confusion_matrix(y_test, predict_dtc)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

clf = RandomForestClassifier(n_estimators=300, criterion='entropy')
clf.fit(x_train, y_train.values.ravel())
#clf.feature_importances_
pred = clf.predict(x_test)
print(accuracy_score(y_test, pred)) 
print("Random Forest AUC : ", roc_auc_score(y_test, pred))
plot_confusion_matrix(y_test, pred) 
cross_val_score(clf, x_train, y_train.values.ravel(), cv=10)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gbm = GradientBoostingClassifier()
gbm.fit(x_train, y_train.values.ravel())
pred = gbm.predict(x_test)
print(accuracy_score(y_test, pred))  
print("GBM AUC : ", roc_auc_score(y_test, pred))
plot_confusion_matrix(y_test, pred) 
cross_val_score(gbm, x_train, y_train.values.ravel(), cv=10)
```

```python
from sklearn.linear_model import LogisticRegressionCV

logistic_regression = LogisticRegressionCV(cv=2)
logistic_regression.fit(x_train, y_train.values.ravel())
predictions = logistic_regression.predict(x_test)
# df1 = pd.DataFrame(predictions)
# df1[0].value_counts()
#print(logistic_regression.score(x_test, y_test))
print(accuracy_score(y_test, predictions)) 
print("ROC AUC : ", roc_auc_score(y_test, predictions))
plot_confusion_matrix(y_test, predictions) 
```
