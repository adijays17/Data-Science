```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('bike.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
col = df.columns.values.tolist()
for e in col:
    # gca stands for 'get current axis'
    ax = plt.gca()
    df.plot(kind='scatter',x='trips',y=e, color='blue', ax=ax)
    plt.show()
```

```python
col = df.columns.values.tolist()
fig, axes = plt.subplots(nrows=2, ncols=2)
for e in col:
    # gca stands for 'get current axis'
    df.plot(kind='scatter',x='trips',y=e, color='blue', ax=axes)
plt.show()
```
